```python
customers = pd.read_csv('../data/customers.csv')
campaigns = pd.read_csv('../data/campaigns.csv')

print("Customers Data")
display(customers)

print("Campaigns Data")
display(campaigns)
```

```python
campaigns = campaigns.rename(columns={
    'cust_id': 'customer_id',
    'customer_name': 'name',
    'email_address': 'email'
})

display(campaigns)
```

```python
campaigns['phone'] = np.nan

display(campaigns)
```

```python
combined = pd.concat([customers, campaigns], ignore_index=True)

display(combined)
```

```python
combined.isnull().sum()
```

```python
combined['phone'] = combined['phone'].fillna('Not Available')

display(combined)
```

```python
combined = combined.drop_duplicates()

display(combined)
```

```python
import re

def validate_email(email):
    
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    
    if re.match(pattern, str(email)):
        return True
    else:
        return False

combined['valid_email'] = combined['email'].apply(validate_email)

display(combined)
```

```python
invalid_records = combined[combined['valid_email'] == False]

display(invalid_records)
```

```python
invalid_records.to_csv(
    '../output/invalid_records.csv',
    index=False
)

print("Invalid Records Saved")
```

```python
clean_data = combined[combined['valid_email'] == True]

clean_data.to_csv(
    '../output/clean_data.csv',
    index=False
)

print("Clean Data Saved")
```

```python
print("Total Records :", len(combined))

print("Valid Records :", len(clean_data))

print("Invalid Records :", len(invalid_records))
```
